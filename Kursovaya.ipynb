```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image,ImageOps
from os import listdir
from os.path import isfile, join
import pandas as pd
```

```python
class_names = ['Футболка/майка', 'Брюки', 'Кофта/свитер', 'Платье', 'Куртка', 
               'Сандалия/туфля', 'Рубашка', 'Кроссовок', 'Сумка', 'Сапог']
```

```python
fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
train_images=np.concatenate((train_images,np.flip(train_images,2)),axis=0)
train_labels=np.concatenate((train_labels,train_labels))
test_images=np.concatenate((test_images,np.flip(test_images,2)),axis=0)
test_labels=np.concatenate((test_labels,test_labels))
```

```python
model1 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(784, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
model1.compile(optimizer="adam", 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model1.fit(train_images, train_labels, epochs=10,validation_split=0.2)
```

```python
test_loss, test_acc = model1.evaluate(test_images, test_labels)
test_acc
```

```python
model2 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(784, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
model2.compile(optimizer="adam", 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model2.fit(train_images, train_labels, epochs=10)
```

```python
test_loss, test_acc = model2.evaluate(test_images, test_labels)
test_acc
```

```python
model1 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(784, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
model1.compile(optimizer="adam", 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model1.fit(train_images, train_labels, epochs=10)
model2 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(784, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.sigmoid)
])
model2.compile(optimizer="adam", 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model2.fit(train_images, train_labels, epochs=10)
```

```python
def opr(filename,model):
    tk=Image.open(filename)
    wpercent = (28 / float(tk.size[0]))
    hsize = int((float(tk.size[1]) * float(wpercent)))
    tk = tk.resize((28, 28), Image.ANTIALIAS)
    m=np.asarray(tk)
    m=np.mean(m, axis=2)
    m=m.astype("uint8")
    m=m/255.0
    m=1-m
    predictions = model.predict(np.asarray([m]))
    rez=[filename]
    for i in range(len(predictions[0])):
        rez.append(round(predictions[0][i]*100,2))
    return rez
```

```python
def f2(directory_name,model):
    z=["Имя файла"]+class_names
    rez=[]
    for i in listdir(directory_name):
        rez.append(opr(join(directory_name,i),model))
    rez=pd.DataFrame(rez)
    rez.columns=z
    return rez
```
